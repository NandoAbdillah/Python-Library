# Merge

```python
df = pd.DataFrame({'Absen' : [1,2,3,4,5], 'Nilai Praktek' : [97,85,92,78,87]})
df
```

```python
df2 = pd.DataFrame({'Absen' : [1,2,3,4,5], 'Nilai PAS' : [77,76,95,97,73]})
df2
```

```python
ndf = pd.merge(df,df2, on='Absen') # Menggabungkan 2 tabel berdasarkan Absen
ndf
```

```python
df3 = pd.DataFrame({'Absen' : [1,2,3,4,5], 'Nilai Akhir' : [0,0,0,0,0]})
ndf = pd.merge(ndf,df3, on='Absen')
ndf
```

```python
ndf['Nilai Akhir'].replace(to_replace=0, value= 90,inplace=True)
```

```python
ndf
```

```python
dt1 = pd.DataFrame({'Absen' : [1,2,3,4,5], 'Nilai Bjawa' :[90,80,65,77,89]})
dt2 = pd.DataFrame({'Absen' : [4,5,6,7,8], 'Nilai Bjerman' : [98,88,89,92,93]})
pd.merge(dt1,dt2) # perhatikan hasil dibawah hanya menyertakan nilai yang sama saja
```

```python
pd.merge(dt1,dt2, how='left') #MEMERGEKAN 2 TBL TANPA TERKECUALI DAN MENGASIGN DATA DENGAN VALUE YANG ADA BILA TIDAK ADA NULLL
# LEFT -> NILAI YANG DIPAKSA MASUK BJAWA
```

```python
pd.merge(dt1,dt2, how='right')
#rRIGHT -> NILAI YANG DIBWA MASUK BJERMAN
```

```python
pd.merge(dt1,dt2, how='outer') # Menggabungkan semua nilai tanpa terkecuali
```

# APPEND

```python
df1 = pd.DataFrame({'Absen' :[1,2,3,4,5], 'UH' :[87,89,86,92,86], 'PAT' : [80,85,96,90,91]})
df2 = pd.DataFrame({'Absen' : [6,7,8,9,10], 'UH' :[85,84,83,91,88], 'PAT' : [90,87,93,91,90]})

print(df1)
print(df2)
```

```python
df1._append(df2) # Namun indeks tidak akan bertambah
```

```python
# kecuali jika kita menuliskan ignore
df1._append(df2, ignore_index = True) # Indeks otomatis bertambah
```

```python
df1._append(df2, ignore_index = True, sort = True) 
```

```python
df1 = pd.DataFrame({'Absen' :[1,2,3,4,5], 'UH' :[87,89,86,92,86], 'PAT' : [80,85,96,90,91]})
df2 = pd.DataFrame({'Absen' : [6,7,8,9,10], 'UH' :[85,84,83,91,88]})
df1._append(df2, ignore_index = True) # Perhatikan hasil patnya
```
