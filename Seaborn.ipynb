```python
# Histogram mewakili penyebaran data dengan
# membentuk bin di sepanjang rentang data
# menggambar batang untuk menunjukkan jumlah pengamatan
```

```python
df.describe()
```

```python
bins = [2,3,4,5,6,7,8,9,10] #Bins digunakan untuk skala sumbu x nya karena diatas kan time_spend_company bentuknya float 
sns.distplot(df['time_spend_company'], bins = bins, 
             kde= True, #KDE digunkan untuk menampilkan sebuah garis/kurva, tulis False untuk menonaktifkannya
            hist = False #HIST digunakan untuk menampilkan histogramnya atau diagram batangnya
            ) 
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins = bins,
             rug = True) #Rug digunakan untuk memberi gap angka pada sumbu x lihat hasil dibawah 
```

```python
sns.distplot(df['time_spend_company'], bins = bins,
             rug = True)
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins = bins,
             rug = True,
             hist_kws = {'color' : 'red', 'edgecolor' : 'blue', 'linewidth': 3, 'alpha' : 0.5 }, #Perhatikan hasilnya dibawah ini untuk batng
             kde_kws = {'color' : 'green', 'linewidth': 4, 'alpha' : 0.3 } #untuk kurva
            ) 
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins = bins,
             rug = True,
             color = 'orange'
            ) 
plt.xticks(bins)
```

# Scatter Plot

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head(10)
```

```python
sns.scatterplot(x= 'age', y= 'fare', data = titanic_df)
```

```python
sns.scatterplot(x= 'age', y= 'fare', data = titanic_df, hue = 'alive')
```

```python
sns.scatterplot(x= 'age', y= 'fare', data = titanic_df, hue = 'alive', style = 'class')
plt.title('Titanic Data Analysis')
```

```python
sns.scatterplot(x= 'age', y= 'fare', data = titanic_df, hue = 'alive', style = 'class', palette = 'gist_rainbow', alpha = 0.5 ) #Alpha digunakan untuk opacity 
plt.title('Titanic Data Analysis')
```

```python
plt.figure(figsize =(12,6))
sns.scatterplot(x ='age', y='fare', data = titanic_df, hue = 'alive', style = 'class', palette = 'inferno')
plt.title('Titanic Data Analysis')
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(x = 'age', y= 'fare', data = titanic_df, hue = 'alive', style = 'class', palette= 'gist_rainbow', alpha = 0.5)
sns.lineplot(x = 'age', y= 'fare', data = titanic_df, hue= 'alive', color = 'green')
plt.title("Titanic Data Analysis")
```
